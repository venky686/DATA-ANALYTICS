```python
import numpy as np
import pandas as pd
data =pd.read_csv('/content/weatherHistory.csv')
print(data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
data['Year'] = data.index.year
data['Month'] = data.index.month
monthly_avg_temp = data.pivot_table(values='Temperature (C)', index='Year', columns='Month', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(monthly_avg_temp, cmap='viridis', annot=True, fmt=".1f", linewidths=.5)
plt.title('Average Monthly Temperature (C) Heatmap')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('/content/weatherHistory.csv')


plt.figure(figsize=(10, 6))
plt.hist(data['Wind Bearing (degrees)'], bins=30, edgecolor='black')
plt.title('Distribution of Wind Bearing (degrees)')
plt.xlabel('Wind Bearing (degrees)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(data['Wind Speed (km/h)'], bins=30, edgecolor='black')
plt.title('Distribution of Wind Speed (km/h)')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Humidity'], bins=30, edgecolor='black')
plt.title('Distribution of Humidity')
plt.xlabel('Humidity')
plt.ylabel('Frequency')
plt.show()
```
