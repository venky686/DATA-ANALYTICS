```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.DataFrame({
    "cust_name":["ram","sham","saram","roy"],
                  "net_speed(mbs)":[2,5,7,1],
                  "pament_failure":[1,0,2,4],
                  "delivery_charges":[25,10,15,30],
                  "purchase_completed":[0,3,2,0]})
print("----net_workdataset----")
print(data)
for i in range (len(data['cust_name'])):
  print(data['cust_name'][i],data['net_speed(mbs)'][i],data['pament_failure'][i],data['delivery_charges'][i],data['purchase_completed'][i])


```

```python
customers = data[data['pament_failure'] > 0]
if not customers.empty:
    print("customers with payment failure:")
    for name in customers['cust_name']:
        print(name)
    plt.stem(customers['pament_failure'], customers['net_speed(mbs)'])
    plt.xlabel('Payment Failure Count')
    plt.ylabel('Net Speed(mbs)')
    plt.title('Internet Diagnosis (Customers with Payment Failure)')
    plt.show()
```

```python
print("---------cust_name---------")
for i in range (len(data['cust_name'])):
  if data['purchase_completed'][i]==0:
    print(data['cust_name'][i])
    plt.title("total abandnment")
    plt.scatter(data['cust_name'][i],data['purchase_completed'][i])


plt.show()

```

```python
print("----root cause----")

root_cause_customers = data[
    (data['net_speed(mbs)'] < 5) &
    (data['pament_failure'] > 0) &
    (data['delivery_charges'] > 20) &
    (data['purchase_completed'] == 0)]

if not root_cause_customers.empty:
  plt.figure(figsize=(7, 5))
  plt.bar(root_cause_customers['cust_name'], root_cause_customers['pament_failure'])
  plt.xlabel('Customer Name')
  plt.ylabel('Payment Failure Count')
  plt.title('Payment Failures for Root Cause Identified Customers')
plt.show()
```
