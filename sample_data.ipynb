```python
#sample data 1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
transaction_amount=[200,3456,6678]
location=["jack_oorc","maream_orke","harry_ores"]
time= pd.to_datetime(['2023-10-26 10:00:00', '2023-10-26 11:30:00', '2023-10-26 14:00:00'])
account_activity=["high","low","high"]
print("unusual transaction")
for i in range (len(transaction_amount)):
  print(transaction_amount[i],location[i],time[i],account_activity[i])
if time[i] > pd.to_datetime('2023-10-26 10:00:00'):
   print("\n",time[i])
elif account_activity[i] == "high":
    print("account_activity",account_activity[i])

```

```python
print("time:")
for i in range (len(time)):
  if time[i] > pd.to_datetime('2023-10-26 10:00:00') and account_activity[i] == 'high':
    print("\n",time[i],"_\n",account_activity[i])
plt.plot(time,account_activity)
plt.show()
```

```python
for i in range (len(transaction_amount)):
  if time[i] > pd.to_datetime('2023-10-26 10:00:00') and account_activity[i] == 'high' and transaction_amount[i]> 2000:
    print(transaction_amount[i],"high risk\n",time[i])
plt.bar(time,transaction_amount)
plt.show()
```

```python
for i in range (len(time)):
 if time[i] > pd.to_datetime('2023-10-26 10:00:00') and account_activity[i] == 'high' and transaction_amount[i]> 2000:
    print("spam:",transaction_amount[i],"high risk",time[i],"location,",location[i])
    print("\n")

plt.pie(transaction_amount,labels=location)
plt.show()
```

```python
#sample data 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
daily_sales=[2000,3456,6678,5623]
product_stock=["mangos","milk","bread","goa"]
expired_date=pd.to_datetime(['2023-10-26 10:00:00', '2023-10-26 11:30:00', '2023-10-26 14:00:00', '2023-10-27 09:00:00'])
seasonal_demand=["summor","common","common","weanter"]
print("unusual transaction")
for i in range (len(daily_sales)):
  print(daily_sales[i],product_stock[i],expired_date[i],seasonal_demand[i])
```

```python
for i in range (len(daily_sales)):
  if expired_date[i] >= pd.to_datetime('2023-10-26 10:00:00') and seasonal_demand[i] == 'common':
    print("fast saling product\n",daily_sales[i],product_stock[i],expired_date[i],seasonal_demand[i])
plt.plot(expired_date,seasonal_demand)
plt.show()
```

```python

if expired_date[i] >= pd.to_datetime('2023-10-26 10:00:00') and seasonal_demand[i] == 'summer' or 'weanter':
    print("slowley saling product\n",daily_sales[i],product_stock[i],expired_date[i],seasonal_demand[i])
plt.bar(expired_date,seasonal_demand)
plt.show()
```

```python
for i in range (len(product_stock)):
  if product_stock[i] == 'summer' or 'weanter' and seasonal_demand[i] != 'common':
    print("out of stock",product_stock[i],"out of stock",seasonal_demand[i])
    print("\n")
plt.pie(daily_sales,labels=product_stock, autopct='%1.1f%%')

plt.show()
```

```python
#sample data 3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
d= pd.DataFrame({
    "order_t":['14:00','27:45','55:25','26:56'],
   "delivery_t":['18:45','24:45','45:30','16:50'],
   "delivery_partner":["rama","sai","tilak","chaten"],
    "customer_rating":[5.5,9.0,9.0,7.0],
    "area":["madanapaiil","businikonda","nimmmapalli","NTR circle"],
    "most_delivery_area":[3,5,6,1]})
print(d)
d['delivery_t']=pd.to_datetime(d['delivery_t'],format='%M:%S')
d['order_t']=pd.to_datetime(d['order_t'],format='%M:%S')
d['time_diff']=d['delivery_t']-d['order_t'].mean()
print("avg_order:",d['time_diff'].mean())
peak_hour=d['order_t'].mean()
print("peak_hour:",peak_hour)
for i in range (len(d['most_delivery_area'])):
  if d['most_delivery_area'][i]> 3 :
    print("most_delivery_area:",d['most_delivery_area'][i],"AREA:",d['area'])
plt.title("analysis")
sns.barplot(x=d['most_delivery_area'],y=d['order_t'])
plt.show()
print("\n")
plt.title("most_delivery_area")
plt.stem(d['most_delivery_area'])
plt.show()
print("\n")
plt.title("delivery_partner")
plt.pie(d['customer_rating'],labels=d['delivery_partner'],autopct='%1.1f%%')
plt.show()
print("\n")
plt.title("customer_rating")
plt.plot(d['order_t'],d['time_diff'],color="blue",alpha=0.5)
plt.show()
```

```python
print("most_deliverd_areA")
for i in range (len(d['customer_rating'])):
  if d['customer_rating'][i]>7.0:
    print(d['customer_rating'][i])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
d= pd.DataFrame({
    'product_purchases':['milk','bread','milk','cloths','bread'],
    'card_adandonment':['phone','clouths','shues','game_pc','tapes'],
    'customer_ratings':[9.1,6.4,7.6,9.2,6.7],
    'payment_methods':['card','cash','card','online_payment','card']

})
print(d)
```

```python
print("\n avg_mostsold item:\n",d['product_purchases'].mode())
plt.title('most sold item')
sns.countplot(x='product_purchases',data=d.mode())
plt.show()
print("high_revenuee_customers\n",d['product_purchases'].mode())
plt.title('high_revenuee_customers')
sns.histplot(x='product_purchases',data=d.mode())
plt.show()
print("mostuse_payment_methods",d['payment_methods'].mode())
plt.title('mostuse_payment_methods')
sns.countplot(x='payment_methods',data=d.mode())
plt.show()
print("customer_ratings",d['customer_ratings'].mode())
plt.title('customer_ratings')
plt.pie(x=d['customer_ratings'], autopct='%1.1f%%')
plt.show()

```
