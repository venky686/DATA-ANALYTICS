```python
import numpy as np
import pandas as pd
data = pd.DataFrame({"name":["raju","sai","venky","chatan"],
"age":[45,67,23,51],"oxy_level":[81,78,89,83],"bed_allocate":[1,2,3,4]})
for i in range (len(data['name'])):
  print(data['name'][i],data['age'][i],data['oxy_level'][i],data['bed_allocate'][i])
print(data)


```

```python
import matplotlib.pyplot as plt
print("priority")
for i in range (len(data['name'])):
  if data['age'][i]>50:
      print(data['age'][i])
      plt.title("age priorityp")
  plt.bar(data['name'][i],data['age'][i])
plt.show()
```

```python
import matplotlib.pyplot as plt

print("pashent helth")
for i in range (len(data['name'])):
  if data['oxy_level'][i]<=80:
    print("pashent condition--",data['name'][i],"--critical")
    print("avalible bed",data['bed_allocate'][i])
  elif data['oxy_level'][i]<=86 :
    print("pashent condition--",data['name'][i],"--observation")
    print("avalible for tempory purpers",data['bed_allocate'][i])
  elif data['oxy_level'][i]>=87 :
    print("pashent condition--",data['name'][i],"--prescription")
    print("bed not rquires",data['bed_allocate'][i])

plt.title("oxy_level priorityp")
plt.bar(data['name'],data['oxy_level'])
plt.show()
```
