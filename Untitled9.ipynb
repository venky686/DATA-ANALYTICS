```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data = pd.DataFrame({
    'Marks':  [70, 80, 67, 69, 78, 92, 65],
    'Study_Time_Hours':[2, 5, 7, 8, 2, 4, 7]
})
print(data)
print("Mean of Marks: ",np.mean(data['Marks']))
print("Median of Marks: ",np.median(data['Marks']))
print("Variance of Marks: ",np.var(data['Marks']))
print("correlation:",np.corrcoef(data['Marks'],data['Study_Time_Hours']))
print("\n")
print("\n")
plt.figure(figsize=(8, 5))
sns.barplot(x=data.index, y='Marks', data=data)
plt.title('Bar Chart of Marks')
plt.xlabel('Student Index')
plt.ylabel('Marks')
plt.show()
print("\n")
print("\n")
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[['Marks', 'Study_Time_Hours']])
plt.title('Boxplot of Marks and Study Time')
plt.ylabel('Value')
plt.show()
print("\n")
print("\n")
mark_counts = data['Marks'].value_counts().sort_index()
plt.figure(figsize=(8, 8))
plt.pie(mark_counts, labels=mark_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Marks Distribution')
plt.axis('equal')
plt.show()
print("\n")
print("\n")
print(data.std())
print("\n")
print("\n")
print(data.describe())
print("\n")
print(data.corr())
print()
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Study_Time_Hours', y='Marks', data=data)
plt.title('Marks vs. Study Time (Scatter Plot)')
plt.xlabel('Study Time (Hours)')
plt.ylabel('Marks')
plt.grid(True)
plt.show()

print("\n")

correlation_matrix = data.corr()

plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Marks and Study Time')
plt.show()
```
