```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
name=["aman","venky","arun","akash","arjun"]
age=[20,31,21,35,19]
work_hower=[1,2.5,3,1.5,2]
calarie_burn=[1000,5000,905,4000,1045]
print("______dataset of fitness______")
for i in range (len(name)):
  print(name[i],age[i],work_hower[i],calarie_burn[i])
avg_work=sum(work_hower)/len(work_hower)
print("average work hour is",avg_work)
hig_cal=max(calarie_burn)
print("highest calorie burn is",hig_cal)
low_workout=min(work_hower)
print("lowest work hour is",low_workout)
median_cal=statistics.median(calarie_burn)
print("median of calorie burn is",median_cal)
std_cal=statistics.stdev(calarie_burn)
print("standard deviation of calorie burn is",std_cal)
index=calarie_burn.index(hig_cal)
print("more active calaries_burn ",name[index])
plt.title("fitness analysis")
plt.bar(name,calarie_burn)
plt.xlabel("members")
plt.ylabel("calaries")
plt.show()
print("\n")
print("\n")
plt.title("fitness analysis")
plt.plot(name,work_hower)
plt.xlabel("members")
plt.ylabel("work hour")
plt.show()
print("\n")
print("\n")
plt.title("fitness analysis")
plt.scatter(name,calarie_burn)
plt.xlabel("members")
plt.ylabel("calaries")
plt.show()
print("\n")
print("\n")
plt.title("fitness analysis")
plt.pie(work_hower,labels=name)
plt.xlabel("members")
plt.ylabel("work hour")
plt.show()
print("\n")
print("\n")
plt.title("fitness analysis")
plt.stem(name,work_hower)
plt.xlabel("members")
plt.ylabel("work hour")
plt.show()
```
