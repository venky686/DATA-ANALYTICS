```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('gym_members_exercise_tracking_synthetic_data.csv')
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Fat_Percentage', y='Calories_Burned')
plt.title('Calories Burned vs. Fat Percentage')
plt.xlabel('Fat Percentage')
plt.ylabel('Calories Burned')
plt.grid(True)
plt.show()
print("\n")
if 'Workout_Frequency (days/week)' in df.columns and 'Water_Intake (liters)' in df.columns and 'Age' in df.columns and 'Workout_Type' in df.columns:
    plt.figure(figsize=(12, 8))
    sns.scatterplot(data=df, x='Water_Intake (liters)',y='Workout_Frequency (days/week)',
                size='Age',sizes=(20, 810),alpha=0.6,hue='Workout_Type')
    plt.title('Water Intake vs. Workout Frequency (Bubble Chart by Age)')
    plt.xlabel('Water_Intake (liters)')
    plt.ylabel('Workout_Frequency (days/week)')
    plt.grid(True)
    plt.legend(title='Workout Type', bbox_to_anchor=(1.07, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
else:
    print("Columns 'Workout_Frequency (days/week)', 'Water_Intake (liters)', 'Age', or 'Workout_Type' not found for bubble chart.")
print("\n")
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Height (m)", y='Max_BPM')
plt.title('BPM Rate and Height (m)')
plt.xlabel('Height (m)')
plt.ylabel('Max_BPM RATE')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("\n")
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
print("\n")
if 'Session_Duration (hours)' in df.columns:
    sample_df = df.head(50).reset_index()
    plt.figure(figsize=(12, 6))
    plt.stem(sample_df.index, sample_df['Session_Duration (hours)'], basefmt="")
    plt.title('Session Duration (Stem Plot - First 50 entries)')
    plt.xlabel('Entry Index')
    plt.ylabel('Session Duration (Hours)')
    plt.grid(True)
    plt.show()
    print("\n")
    print("\n")
else:
    print("Column 'Session_Duration (hours)' not found for stem plot.")
if 'Workout_Type' in df.columns:
    workout_type_counts = df['Workout_Type'].value_counts()
    plt.figure(figsize=(10, 8))
    plt.bar(workout_type_counts.index, workout_type_counts.values)
    plt.title('Distribution of Workout Types')
    plt.show()
else:
    print("Column 'Workout_Type' not found for bar chart.")
```
