```python
import numpy as np
import pandas as pd
data=pd.read_csv('/content/cleaned_global_water_consumption.csv')
print(data.head())
```

```python
import matplotlib.pyplot as plt
avg_Agricultural_Water_Use  =data.groupby('Year')['Agricultural Water Use (%)'].mean()
print(f"Average Agricultural Water Use: {avg_Agricultural_Water_Use} Liters per Day")
plt.title('Average Agricultural Water Use')
plt.stem(avg_Agricultural_Water_Use.index, avg_Agricultural_Water_Use)
plt.xlabel('Year')
plt.ylabel('Average Agricultural Water Use')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(data['Year'], data['Total Water Consumption (Billion Cubic Meters)'])
plt.xlabel('Year')
plt.ylabel('Total Water Consumption (Billion Cubic Meters)')
plt.title('Total Water Consumption Over Time')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title("Industrial Water Use (%)")
plt.bar(data['Year'], data['Industrial Water Use (%)'])
plt.xlabel("year")
plt.ylabel("Industrial Water Use ")
plt.show()
```

```python
sum_Industrial_Water_Use =data.groupby('Country')['Industrial Water Use (%)'].sum()
print(sum_Industrial_Water_Use)
plt.figure(figsize=(12, 6))
plt.pie(sum_Industrial_Water_Use, labels=sum_Industrial_Water_Use.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Industrial Water Use by Country')
plt.show()
```

```python
def tank(capacity_value):
  t = 3000
  if capacity_value < t:
    print("tank : ", capacity_value)
    return False
  else:
    print(" tank is empty")
    return True

while True:
  try:
    user_input_str = input("Enter the tank capacity (enter 0 to stop): ")
    user_input = int(user_input_str)

    if user_input == 0:
      print("Stopping the program.")
      break

    if tank(user_input):
      print("Stopping the program as tank is empty/full.")
      break
  except ValueError:
    print("Invalid input. Please enter an integer.")
```
