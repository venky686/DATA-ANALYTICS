Loops & Conditionals

```python
age=int(input("Enter age: "))

if age > 18 and age < 50:
  print("Allowed")
else:
  print("Not allowed")
```

```python
price=int(input("Enter price: "))
if price < 2500:
  print("Nothing")
elif price > 2500 and price < 5000:
  print("You got 500 discount")
  print("Price:",price)
  print("Discounted Price:",price-500)
else:
  print("You got 1000 discount")
  print("Price:",price)
  print("Discounted Price:",price-1000)
```

```python
l=[10,20,30,40,50]
for i in l:
  print(i)
```

```python
student = {"name": "Karthik","age": 21,"branch": "CSE"}
for key in student:
    print(key)
```

```python
for value in student.values():
    print(value)
```

```python
for key, value in student.items():
    print(key, ":", value)
```

Comprehensions

```python
numbers = [1, 2, 3, 4, 5]
squares = [x**2 for x in numbers]
print(squares)
```

```python
numbers = [1, 2, 3, 4, 5, 6]
even = [x for x in numbers if x % 2 == 0]
print(even)
```

```python
numbers = [1, 2, 3, 4, 5]
square_dict = {x: x**2 for x in numbers}
print(square_dict)
```

```python
numbers = range(1, 11)
even_square = {x: x**2 for x in numbers if x % 2 == 0}
print(even_square)
```

```python
numbers = [1, 2, 2, 3, 4, 4, 5]
unique = {x for x in numbers}
print(unique)
```
